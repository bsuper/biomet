```python
%pylab inline
import matplotlib.pyplot as plt
import scipy.io as sio
import numpy as np
import datetime
from sklearn import cross_validation, svm, linear_model, preprocessing

def load_mat_file(filename):
    return sio.loadmat(filename)

def convert_doy_to_yyyymmdd(year, doy):
    return datetime.date(year,1,1) + datetime.timedelta(doy - 1)

def get_yyyymmdd(year, doy):
    yyyymmdd = []
    for i in range(year.shape[0]):
        dt = convert_doy_to_yyyymmdd(year[i][0], doy[i][0])
        yyyymmdd.append(dt)
    return yyyymmdd
```

```python
DATA_FILE = "WP_2012195to2015126_L3.mat"

# data
data = load_mat_file(DATA_FILE)
# general flux data
d = data['data'][0][0]
# metdata
m = data['Metdata'][0][0]

# air temperature
air_temp = m['TA']
# water table depth
water_table = m['WT']
# water table level gap filled
wt_gf = d['wt_gf']
# co2 gap filled
co2_gf = d['wc_gf']
# ch4 gap filled
ch4_gf = d['wm_gf']
# hour of day
hod = d['time']
# day of year
doy = d['DOY']
# year
year = d['year']
# yyyymmdd datetime
yyyymmdd = get_yyyymmdd(year, doy)
```
